```python
# libraries
import os
import pandas as pd
```

```python
def data_preprocessing (csv_data):
    
    # features in dataset
    rows_csv_data, columns_csv_data = csv_data.shape

    # selecting features
    selected_columns = 120

    # preprocessing
    i = 0
    extracted_data = pd.DataFrame()

    while (i < (columns_csv_data - 2)):
        
        # extracting features from laser data
        if (i < (columns_csv_data - 14)):
            extracted_data = pd.concat([extracted_data, csv_data.iloc[: , i : (i + selected_columns)].mean(axis = 1)], axis = 1)
            i += selected_columns

         # extracting features from position data of goal and current pose
        else:
            extracted_data = pd.concat([extracted_data, csv_data.iloc[: , i : (i + 2)]], axis = 1)
            extracted_data = pd.concat([extracted_data, csv_data.iloc[: , (i + 2) : (i + 4)].sum(axis = 1)], axis = 1)
            i += 4

    extracted_data = pd.concat([extracted_data, csv_data.iloc[: , (columns_csv_data - 2) : columns_csv_data]], axis = 1)
    return extracted_data
```

```python
if __name__ == '__main__':
    
    # removing existing extracted file
    output_file = 'Training_set_merged.csv'
    folder_name = 'Features'

    if(os.path.exists(output_file) and os.path.isfile(output_file)):
        os.remove(output_file)
        print("file deleted")
    else:
        print("file not found")
    
    # The file named 'July22_69.csv' has been considered
    # reading the list of CSVs in the training folder
    CSV_list = os.listdir(os.path.realpath('') + '\\' + folder_name)
    
    # reading all the csv files in the path, preprocessing it
    extracted_data = pd.DataFrame()
    
    # storing extracted data
    temp = 0
    for CSV_file in CSV_list:
        csv_data = pd.read_csv(folder_name + '/' + CSV_file, header = None)
        extracted_data = data_preprocessing(csv_data)
        extracted_data.to_csv(output_file, mode='a', header = None, index=False)
```
