```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# function to plot loss
def plot_loss(history):
    plt.plot(history.history['loss'], label='E_in Validation')
    plt.xlabel('Iterations')
    plt.ylabel('E_in')
    plt.legend()
    plt.grid(True)
```

```python
if __name__ == '__main__':

  # Reading a data file
  data = pd.read_csv('Training_set_merged.csv', header = None)
  rows, columns = data.shape
  Y_val = data.iloc[:, columns - 2 : columns]
  X_val = data.iloc[:, 0 : columns - 2]

  # data normalization
  normalizer = tf.keras.layers.Normalization(axis=-1)
  normalizer.adapt(np.array(X_val))

  # Linear regression model using Keras
  linear_model = tf.keras.Sequential([ normalizer, layers.Dense(units=1)])
  linear_model.predict(X_val)
  linear_model.layers[1].kernel
  linear_model.compile( optimizer=tf.optimizers.Adam(learning_rate=0.01), loss='mean_absolute_error')

  # Training the model
  history = linear_model.fit(
      X_val,
      Y_val,
      epochs=100,
      verbose=0)

  # Loss using model
  plot_loss(history)

  # results
  test_results = {}
  test_results['linear_model'] = linear_model.evaluate(X_val, Y_val, verbose=0)

  print("\nThe mean squared error for the linear regression: ", "{:.5f}".format(test_results['linear_model']), "\n")
```
